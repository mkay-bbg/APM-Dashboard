```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('data_v1.1.csv')

# Data wrangling
df['datetime'] = pd.to_datetime(df['datetime'])
df["time"] = df['datetime'].dt.time
df["date"] = df['datetime'].dt.date
df['Net Change'] = df['ppl_in'] - df['ppl_out']

df_grouped = df.groupby(['datetime', 'date', 'time', 'Car', 'door']).sum()

Capacity_per_car = 76
df_grouped['Ppl Onboard'] = df_grouped['Net Change'].cumsum()
df_grouped['Occupancy Rate'] = df_grouped['Ppl Onboard'] / Capacity_per_car

df_grouped
```

```python
df_grouped_datetime = df.groupby(by="datetime").sum()[["ppl_out", "ppl_in", "Net Change"]]

plt_all = px.line(df_grouped_datetime, x=df_grouped_datetime.index.get_level_values("datetime"), y=["ppl_out", "ppl_in"])
plt_all.add_bar(x=df_grouped_datetime.index.get_level_values("datetime"), y=df_grouped_datetime["Net Change"], name="Net people flow")

# buttons for timeframe selection
plt_all.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1h",
                     step="hour",
                     stepmode="backward"),    
                dict(count=1,
                     label="1d",
                     step="day",
                     stepmode="backward"), 
                dict(count=1,
                     label="all",
                     step="all",
                     stepmode="backward")                        
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
```
